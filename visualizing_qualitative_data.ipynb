```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listing = pd.read_csv('listings.csv')
```

```python
listing
```

```python
sn.countplot(x = 'neighbourhood_group', data = listing)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listing)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listing)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listing, ci = False)
plt.show()
```

```python
sn.barplot(x='neighbourhood_group', y='price', data=listing)
plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.show()
```

```python
sn.boxplot(x='neighbourhood_group', y='price', data=listing)
plt.title("Price Distribution by Neighbourhood Group")
plt.show()
```

```python
sn.stripplot(x='neighbourhood_group', y='price', data=listing, jitter=True)
plt.title("Prices by Neighbourhood Group")
plt.show()
```

```python
pivot_table = listing.pivot_table(index='neighbourhood_group', values='price', aggfunc='mean')
sn.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title("Heatmap of Mean Prices by Neighbourhood Group")
plt.show()
```

```python
sn.histplot(data=listing, x='price', hue='neighbourhood_group', kde=True)
plt.title("Price Distribution by Neighbourhood Group")
plt.show()
```

```python
listing['neighbourhood_group'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6))
plt.title("Proportion of Neighbourhood Groups")
plt.ylabel("")
plt.show()
```
